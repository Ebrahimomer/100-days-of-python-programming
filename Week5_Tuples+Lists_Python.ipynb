````python
## 5Ô∏è‚É£ Tuples ‚Äî Immutable Lists
# What Are Tuples in Python?

# A tuple is a type of sequence in Python ‚Äî just like a list ‚Äî but with one important difference:

# Tuples cannot be changed once created.

# This means they are immutable.
# Tuples:

# - ordered
# - **cannot change**
# - use **( )**
# my_tuple = (10, 20, 30)
# You can store different types inside a tuple:
# info = ("Alice", 25, True)
# Tuples Are Immutable

# Once a tuple is created:

# ‚ùå You cannot:

# change its values

# add new items

# remove items

# Example:
t = (1, 2, 3)
t[0] = 10   # ‚ùå This gives an error

````

````python
# Converting Between Lists and Tuples

# You can convert a tuple to a list if you need to modify it.

# Tuple ‚Üí List:
t = (1, 2, 3)
l = list(t)   # Now it's a list

# List ‚Üí Tuple:
l = [1, 2, 3]
t = tuple(l)

# When to Use Tuples vs Lists
# Use Tuple when‚Ä¶	                             Use List when‚Ä¶
# Data should not change	                 Data will change over time
# You need faster performance	             You need to add/remove items
# You want to protect data	                You want flexibility
# You‚Äôre using it as a dictionary           key	Data is dynamic
# In simple words:

# A tuple is like a ‚Äúlocked‚Äù list.
# Once you create it, it stays the same forever.
# This makes your data safe, clean, and reliable.
````
